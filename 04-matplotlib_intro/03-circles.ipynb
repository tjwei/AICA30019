```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(-1,1, 100)
```

```python
plt.plot(x, np.sqrt(1-x**2))
plt.plot(x, -np.sqrt(1-x**2))
```

```python
t = np.linspace(0, 2* np.pi, 100)
```

```python
x = np.cos(t)
y = np.sin(t)
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,5))
t = np.linspace(0, 2* np.pi, 6)
plt.plot(np.cos(t), np.sin(t))
```

```python
plt.figure(figsize=(5,5))
t = np.linspace(0, 4* np.pi, 6)
plt.plot(np.cos(t), np.sin(t))
```

```python
plt.figure(figsize=(5,5))
t = np.linspace(0, 2* np.pi, 11)
r = np.array([2,1]*5+[2])
plt.plot(r*np.cos(t), r*np.sin(t))
```

```python
plt.figure(figsize=(5,5))
t = np.linspace(0, 2* np.pi, 11)
r = np.arange(11)%2+1
plt.plot(r*np.cos(t-0.3), r*np.sin(t-0.3))

t = np.linspace(0, 2* np.pi, 101)
r = 2
plt.plot(r*np.cos(t), r*np.sin(t))


t = np.linspace(0, 2* np.pi, 5)
r = 2
plt.plot(r*np.cos(t), r*np.sin(t))
plt.savefig('magic circle.png')
```
