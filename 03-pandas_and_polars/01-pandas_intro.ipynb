```python
m_f_ratio_by_school = {}
for school_name in df["學校名稱"].unique():
    series_total = df[df["學校名稱"] == school_name][["男生計", "女生計"]].sum()
    m_f_ratio_by_school[school_name] = series_total['男生計']/series_total['女生計']
df_by_loop = pd.DataFrame.from_records(list(m_f_ratio_by_school.items()), 
                          columns=['school_name', 'm_f_ratio'])
```

```python
gb = df[["學校名稱", '男生計', '女生計']].groupby(["學校名稱"])
df_by_school = gb.sum()
df_by_school['m_f_ratio'] = df_by_school['男生計'] / df_by_school['女生計']
df_by_school
```

```python
gb = df[["學校名稱", '男生計', '女生計']].groupby(["學校名稱"])
df_by_school = gb.sum()
df_by_school['m_f_ratio'] = df_by_school.eval('男生計 / 女生計')
df_by_school
```
