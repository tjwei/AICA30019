```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((600, 800, 3), dtype=np.float)
plt.imshow(img)
```

```python
img[200:300, 400:500] = (0,1,0)
plt.imshow(img)
```

```python
y, x = np.indices((600, 800))
img[(x-450)**2 + (y-350)**2 < 50**2] = (1,0,0)
plt.imshow(img, interpolation='bilinear')
```

```python
plt.imshow(img[200:400, 300:500], interpolation='bilinear')
```

```python
img[250-10:250+10, 300:400] = (0,0,1)
img[200:300, 350-10:350+10] = (0,0,1)
plt.imshow(img[200:400, 300:500], interpolation='bilinear')
```

```python
plt.imshow(img, interpolation='bilinear')
```

```python
box = (x<400) & (x>=300) & (y>=300) & (y<=400)
img[box]  = 0
line1 = y>=350-0.5*(x-300)
line2 = y<=350+1.*(x-300)
line3 = y<=400-2.*(x-350)

img[line1 & line2 & line3 & box] = 1
plt.imshow(img, interpolation='bilinear')
```

```python
img[0:100, 0:100] = img[300:400, 400:500]
img[0:100, 100:200] = img[200:300, 300:400]
img[0:100, 200:300] = img[250:350, 350:450]
img[0:100, 300:400] = img[300:400, 400:300:-1]
img[0:100, 400:500] = img[400:300:-1, 300:400]
img[0:100, 500:600] = img[400:300:-1, 400:300:-1]
img[0:100, 600:700] = img[200:400:2, 300:500:2]
img[100:300:2, 0:200:2] = img[300:400, 300:400]
# 貼四個圖形
img[101:300:2, 0:200:2] = img[200:300, 300:400]
img[100:300:2, 1:200:2] = img[300:400, 400:500]
img[101:300:2, 1:200:2] = img[200:300, 400:500]
# 行列互換 hint:(keyword transpose, .T)
img[300:500, 0:200]=np.transpose(img[200:400, 300:500], axes=[1,0,2])
# 旋轉九十度 
img[100:300, 500:700]=img[300:500, 0:200][:, ::-1]
# 綠色的圓圈
img[100:200, 200:300, 1]  = img[300:400, 400:500, 0]
img[100:200, 300:400, :2] = img[300:400, 300:400, :2]
plt.imshow(img, interpolation='bilinear')
```

```python
img[...]  = 1
img[300:450, 100:700] = (1,0,0)
cx,cy = 250, 450
img[(x-cx)**2+(y-cy)**2<80**2]=1
img[(x-cx)**2+(y-cy)**2<70**2]=0
img[(x-cx)**2+(y-cy)**2<50**2]=0.5
img[(x-cx)**2+(y-cy)**2<20**2]=(0,0,1)

cx,cy = 550, 450
img[(x-cx)**2+(y-cy)**2<80**2]=1
img[(x-cx)**2+(y-cy)**2<70**2]=0
img[(x-cx)**2+(y-cy)**2<50**2]=0.5
img[(x-cx)**2+(y-cy)**2<20**2]=(0,0,1)
img[200:300, 500:600] = 1
yy,xx = np.indices((100,100))
img[200:300, 500:600][yy>xx] = (1,0,0)

plt.imshow(img, interpolation='bilinear')
```

```python
img = np.zeros((800,800,3))
y, x = np.indices(img.shape[:2])
cx,cy = 300, 300
circle1 = (x-cx)**2+(y-cy)**2<200**2
cx,cy = 500, 500
circle2 = (x-cx)**2+(y-cy)**2<200**2
cx,cy = 300, 500
circle3 = (x-cx)**2+(y-cy)**2<200**2
cx,cy = 500, 300
circle4 = (x-cx)**2+(y-cy)**2<200**2
img[circle1]+=(.2,0,0)
img[circle2]+=(.2,0,0)
img[circle3]+=(.2,0,0)
img[circle4]+=(.2,0,0)
img[circle1 & circle2 & circle3 & circle4] =(0,0,1)
plt.imshow(img, interpolation='bilinear')
```

```python
img[:, :, 2]  = 1-y/600
img[:, :,  0]  = 0.5+y/600*0.5
img[:, :,  1]  = 0.5+y/600*0.5

img[300:450, 100:700] = (1,0,0)
cx,cy = 250, 450
img[(x-cx)**2+(y-cy)**2<80**2]=1
img[(x-cx)**2+(y-cy)**2<70**2]=0
img[(x-cx)**2+(y-cy)**2<50**2]=0.5
img[(x-cx)**2+(y-cy)**2<20**2]=(0,0,1)

cx,cy = 550, 450
img[(x-cx)**2+(y-cy)**2<80**2]=1
img[(x-cx)**2+(y-cy)**2<70**2]=0
img[(x-cx)**2+(y-cy)**2<50**2]=0.5
img[(x-cx)**2+(y-cy)**2<20**2]=(0,0,1)

y_bound = (y>200) & (y<300)
x_bound = abs(x-400)< y-100
img[x_bound & y_bound] = (1,0,0)
y_bound = (y>220) & (y<280)
x_bound = abs(x-400)< y-120
img[x_bound & y_bound] = 0
img[200:300, 390:410]=(1,0,0)
plt.imshow(img)
```

```python
img[...] = 0
sine_wave = np.floor(300+200*np.sin(x/800*2*np.pi))

img[ (y<sine_wave+2) &  (y>sine_wave-2)] = 1
plt.imshow(img, interpolation="bilinear")
```

```python
img[...] = 0
sine_wave = np.floor(300+200*np.sin(x/800*2*np.pi))

img[ (y<sine_wave) &  (y >300)] = (1,0,0)
img[ (y>sine_wave) &  (y <300)] = (0,0,1)
plt.imshow(img, interpolation="bilinear")
```

```python
img[...] = 0
curve1 = y<np.floor(500-0.001*(x-400)**2)
curve2 = y>np.floor(400-0.0005*(x-400)**2)

img[curve1 & curve2] = (1,0,0)

plt.imshow(img, interpolation="bilinear")
```

```python
img[...] = 0
line1 = y - (400-0.7*x)
line2 = y - (400+0.7*(x-800))
img[line1 * line2 > -5000] =1

plt.imshow(img, interpolation="bilinear")
```

```python
img[...]=0
X = (x-400)/200
Y = -(y-300)/200
img[(X**2+Y**2-1)**3-X**2 * Y**3<0]= (1,0,0)

plt.imshow(img, interpolation="bilinear")
```

```python
img[...]=0
X = (x-400)/30
Y = -(y-300)/30
heart = X**2+(Y-(2*(X**2+np.abs(X)-6))/(3*(X**2+np.abs(X)+2)))**2<36 
img[heart]= (1,0,0)

plt.imshow(img, interpolation="bilinear")
```

```python
img[...]=0
X = (x-400)/30
Y = -(y-300)/30
R = np.sqrt(X**2 + Y**2)
t = np.arctan2(Y,X)
img[R<5]= (1,0,0)
img[(t > 0) & (t<3.14/4)]=(0,0,1)

plt.imshow(img, interpolation="bilinear")
```
