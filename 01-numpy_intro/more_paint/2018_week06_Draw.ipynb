```
import numpy as np
img = np.zeros( (600, 800, 3), dtype=np.float)
```

```
#%matplotlib inline
import matplotlib.pyplot as plt
```

```
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
for y in range(0,600, 100):
  for x in range(0, 800, 100):
    if (x+y)%200 == 100:
      img[y:y+100,x:x+100]=(0.5, 0.1, 1)
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
img[:, 350:450]=(1,0,0)
img[150:250, 150:650]=(1,0,0)
plt.imshow(img)
```

```
img = np.zeros( (600, 800, 3), dtype=np.float)
for y in range(600):
  for x in range(800):
    r2 = (x-400)**2 + (y-300)**2
    if  r2 < 300**2:
      img[y,x]=(0, 0 , (r2**0.5-200)/100)
    if r2 < 200**2:
      img[y,x]=(0, (r2**0.5-100)/100 , 0)
    if r2 < 100**2:
      img[y,x]=(r2**0.5/100, 0 , 0) 
plt.imshow(img)
```

```
x, y = np.meshgrid(range(800), range(600))
img = np.zeros( (600, 800, 3), dtype=np.float)
img[0.7*(x-400)**2  + 2*(y-300)**2 <= 200**2] = (1,0,0)
plt.imshow(img)
```

```
x, y = np.meshgrid(range(800), range(600))
img = np.zeros( (600, 800, 3), dtype=np.float)
line1 = x+y>=400
line2 = x-2*y < 100
line3 = y < 400 -0.1*x
img[line1 & line2 & line3] = (1,0,0)
plt.imshow(img)
```
