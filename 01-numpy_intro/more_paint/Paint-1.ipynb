```python
for i in range(100, 300):
    y = x = i
    a[y, x, 0] =255
    a[y, x, 1] =255
    a[y, x, 2] =255
Image.fromarray(a)
```

```python
for i in range(100, 300):
    y = x = i
    for j in range(-20,21):
        a[y+j, x, 0] = 0
        a[y+j, x, 1] = 0 
        a[y+j, x, 2] = 0
Image.fromarray(a)
```

```python
a[:,:]=(255,0,0)
Image.fromarray(a)
```

```python
cx, cy = 300, 400
r = 250
for x in range(800):
    for y in range(600):
        if (x-cx)**2 + (y-cy)**2 <= r **2:
            a[y, x] = (0, 255, 255)
Image.fromarray(a)
```
