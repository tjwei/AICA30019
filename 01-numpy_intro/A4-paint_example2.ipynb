```python
y, x=np.ogrid[-1:1:500j,-1.5:0.5:500j]
c = x+1j*y
z = 0.
N = 500
img = zeros((500, 500))
for g in range(N):
    mask = np.abs(z) < 2
    img = img + 1./N*mask
    z = z**2+c
imshow((1-img))
```
